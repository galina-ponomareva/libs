```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
rcParams['figure.figsize'] = 6, 4.5
```

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

```python
fig, ax= plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1, label='x^2')
ax1.set_title('График 1')
ax1.set_xlim([-5, 5])
ax1.legend(loc='lower right', fontsize=10, frameon=False)

ax2.plot(x, y2, label='2*x + 0.5')
ax2.set_title('График 2')
ax2.legend(loc='lower left', fontsize=10, frameon=False)

ax3.plot(x, y3, label='-3*x - 1.5')
ax3.set_title('График 3')
ax3.legend(loc='lower right', fontsize=10, frameon=False)

ax4.plot(x, y4, label='sin(x)')
ax4.set_title('График 4')
ax4.legend(loc='lower right', fontsize=10, frameon=False)

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
