```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

plt.style.use('fivethirtyeight')
rcParams['figure.figsize'] = 6, 4.5
```

```python
df = pd.read_csv('creditcard.csv', sep=',')
df.head(4)
```

```python
index = [1, 2]
plt.bar(index, df['Class'].value_counts())
plt.show()
```

```python
plt.bar(index, df['Class'].value_counts())
plt.yscale(value="log")
plt.show()
```

```python
v1_deal = df.loc[(df['Class'] == 0), 'V1']
v1_deal.min()
```

```python
v1_fool = df.loc[(df['Class'] == 1), 'V1']
v1_fool.head()
```

```python
plt.hist(v1_deal, bins=20, density=True, alpha=0.5, color='Grey', label='Class 0')
plt.hist(v1_fool, bins=20, density=True, alpha=0.5, color='Red', label='Class 1')
plt.yscale(value="log")

title_dict = {'fontsize': '16', 'fontweight': 'bold', 'color': 'DarkBlue', 'family': 'Calibri'}
lable_dict = {'fontsize': '12', 'color': 'DarkBlue', 'family': 'Calibri'}

plt.title('Bank operations', fontdict=title_dict)
plt.xlabel('Class', fontdict=lable_dict)
plt.ylabel('Number of operations', fontdict=lable_dict)
plt.legend(loc='upper left', fontsize=10, frameon=False)

plt.show()
```
